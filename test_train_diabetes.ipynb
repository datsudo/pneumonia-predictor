# Training Test

Using Diabetes dataset


```python
from sklearn.model_selection import train_test_split

from pneumonia_predictor.backend.data_fetcher import load_data
from pneumonia_predictor.backend.rf_active_smote import RfActiveSMOTE
from pneumonia_predictor.backend.utils import get_feature_target_set
```

## Data Splitting

```python
dataset = load_data("diabetes")
```

```python
X, y = get_feature_target_set(dataset, "Diabetes_binary")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## Training

```python
rf_active_smote = RfActiveSMOTE(
    X_train,
    y_train,
    X_test,
    y_test,
    "Diabetes_binary",
    sampling_ratio=0.040607994762898406,
    update_sampratio_per_iter=True,
)
```

```python
rf_active_smote.train(n_iterations=10)
```
